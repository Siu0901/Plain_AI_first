```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
key = "인공지능"
url = f"https://www.ohmynews.com/NWS_Web/Search/s_ohmynews_sub.aspx?keyword={key}&pv=news"
```

```python
respons = requests.get(url)

html = respons.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
newsb = soup.select('div.content_box.dbt40.clearfx')[0]
newsb
```

```python
news_list = newsb.find_all("div",{"class":"text_area"})
news_list[0].get_text()
```

```python
title_list = []
body_list = []
info_list = []
category_list = []
for news in news_list:

  # 기사 제목 추출
  title = news.find("strong").get_text()
  title_list.append(title)

  # 기사 내용 일부 추출 (여긴 서브 타이틀이긴 한데 걍 기사 내용이라 치자)
  try:
    body = news.select("div.sb_tt")[0].get_text()
    body_list.append(body)
  except:
    body_list.append("")

  # info 추출
  info = news.select("div.information")[0].get_text()
  info = [i.strip() for i in info.split('ㅣ')] # 아니 저거 짝대기 |이게 아님 ㅅㅂ
  info_list.append(info)

  # category 추출
  category = news.select("div.category")[0].get_text().strip()
  category_list.append(category)

index = 8
print(title_list[index])
print(body_list[index])
print(info_list[index])
print(category_list[index])
```
