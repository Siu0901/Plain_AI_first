```python
import requests
from bs4 import BeautifulSoup
```

```python
keyword, page = "삼성전자", 1 # 이걸 하는 이유는 지금은 동적 크롤링이 아님

url = f"https://search.hankyung.com/search/news?query={keyword}&page={page}"
```

```python
response = requests.get(url) # get 함수 이용해서 페이지 가져오기

html = response.text # body
soup = BeautifulSoup(html, "html.parser") # html parser를 이용해서 필요한 정보만 가져오기
```

- find() => 일치하는 태그가 여러 개 있다면 가장 첫 번째의 태그를 가져오는 함수
  - find("태그 이름", {"속성": "속성 값"})

- find_all() => 일치하는 태그가 여러 개 있다면 전부 가져오는 함수
  - find_all("태그 이름", {"속성": "속성 값"})

- select() => CSS 선택자를 활용해 조건이 일치하는 태그를 가져오는 함수

```python
news_body = soup.select("div.section.hk_news")[0] #select문은 결과를 리스트로 반환한다
news_body
```

```python
news_list = news_body.find_all("div", {"class": "txt_wrap"}) #예도 리스트 반환임
for news in news_list:
  print(news.get_text()) # get_text()는 html코드 다 제거하고 순수 텍스트만 출력
```

```python
title_list = []
body_list = []
info_list = []
hash_list = []
for news in news_list:
  # 제목 가져오기
  title = news.find("em").get_text()
  title_list.append(title)

  # 본문 미리보기 가져오기
  description = news.select("p.txt")[0].get_text() #여기 부분 나중에 질문하자 ㅅㅂ
  body_list.append(description)

  # info 가져오기
  info = news.select("p.info")[0].get_text()
  info = [i.strip() for i in info.split('|')] # 아니 승준 센빠이는 사람인것인가
  # 저 주석 코드는 내 원시 코든데 비효율적임
  #.replace("\n", "").replace("\r","").replace("|","").split()
  # info[1] = info[1] + " " + info.pop(2)
  info_list.append(info)

  # hash_wrap 가져오기
  try:
    hash_wrap = news.select("div.hash_wrap")[0].get_text().strip().split('\n')
    # 이 주석 코드도 내 원시 코드임. 승준 선배는 아니 진짜 사람이신가
    # hash_wrap = hash_wrap[1:-1]
    hash_list.append(hash_wrap)
  except:
    hash_list.append([])

i = 9
print(title_list[i])
print(body_list[i])
print(info_list[i])
print(hash_list[i])
# info, hash_wrap 가져오고 직접 값 출력해보기
# to-do
```
